```python
plt.scatter(df["Age"],df["Salary"],c=df["Dept"].map({"HR":"yellow","IT":"red","Farmer":"orange", "Finance": "green", "Marketing": "blue", "Management": "purple"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs SAlary vs Exp")
plt.legend()
plt.show()
```

```python
color = {"HR":"yellow","IT":"red","Farmer":"orange", "Finance": "green", "Marketing": "blue", "Management": "purple"}
for Dept,color in color.items():
  df_Dept = df[df["Dept"]==Dept]
  plt.scatter(df_Dept["Age"],df_Dept["Salary"],c=color,label=Dept)

plt.legend()
```

```python
#(OO) API
fig , axs=plt.subplots(1,3, figsize=(15,10))

#LINE PLOT

axs[0].plot(sort_age["Age"],df["Salary"],color ="red",marker="o",linewidth="2",markersize=2)
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")
axs[0].grid()
axs[0].set_title("Line Plot")

#histogram

axs[1].hist(df["Salary"],bins=5,color="green")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")
axs[1].grid()
axs[1].set_title("Histogram")

#box plot

axs[2].boxplot(df["Salary"])
axs[2].grid()
axs[2].set_title("Box Plot")

plt.tight_layout()

plt.savefig("plot.png")
plt.show()
```

```python
import pandas as pd

data2={
"Year": [2020, 2021, 2022, 2023, 2024],
"Sales": [50000, 60000, 75000, 90000, 105000],
"Expense": [30000, 35000, 40000, 50000, 60000],
"Profit": [20000, 25000, 35000, 40000, 45000]
}

```

```python
df2=pd.DataFrame(data2)
df2
```

```python
plt.plot(df2["Year"],df2["Sales"],label="Sales")
plt.plot(df2["Year"],df2["Expense"],label="Expense")
plt.plot(df2["Year"],df2["Profit"],label="Profit")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Financial Analysis")
plt.grid()
plt.show()
```

```python
ax=plt.axes(projection="3d")
ax.scatter(df2["Year"],df2["Sales"],df2["Expense"])
ax.set_xlabel("Year")
ax.set_ylabel("Sales")
ax.set_zlabel("Expense")
plt.show()
```

```python
import plotly.express as px
```

```python
fig=px.line(df2,x="Year",y=["Sales","Expense","Profit"],title="Financial Analysis")
fig.show()
```

```python
fig=px.scatter_3d(df2,x="Year",y="Sales",z="Expense",
                  title="3D scatter plot")
fig.show()
```
