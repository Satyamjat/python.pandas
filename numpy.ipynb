```python
#multiply matrix
a=np.ones((2,3))
b=np.full((3,2),2)
print(a)
print(b)

c=np.matmul(a,b)
print(c)
```

```python
# reshape
before=np.array([[1,2,3,4],[4,5,6,7]])
print(before)
after=before.reshape((2,2,2))
print(after)
```

```python
# vertical stack
v1=np.array([2,34,5,6,2])
v2=np.array([3,4,5,6,45])
V=np.vstack([v1,v2,v2,v1])
print(V)

```

```python
#horizontal stack
h1=np.array([2,34,5,6,2])
h2=np.array([3,4,5,6,45])
H=np.hstack((h1,h2,h2))
print(H)
```
