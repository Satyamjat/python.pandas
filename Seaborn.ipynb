```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.set_style('ticks')
sns.set_context('paper')

sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",
                hue="island",palette='Dark2')
sns.despine()
```

```python
# Here are the available options for sns.set_style() and sns.set_context():

# sns.set_style() options: These control the aesthetic style of the plots.

# 'darkgrid' (default)
# 'whitegrid'
# 'dark'
# 'white'
# 'ticks'
# sns.set_context() options: These scale the plot elements to make them more
# readable in different contexts (e.g., presentations, papers).

# 'paper'
# 'notebook' (default)
# 'talk'
# 'poster'
# You can also adjust the font scale for set_context() using the font_scale
# parameter, for example: sns.set_context('talk', font_scale=1.25).

```

```python
sns.set_context("talk")
sns.swarmplot(data=penguins, x="species",
              y="body_mass_g", hue="island")
sns.despine( )
```

```python
sns.set_context("paper")
sns.histplot(data=penguins, x="body_mass_g", hue="sex",
             multiple="stack")
```

```python
sns.regplot(data=penguins,x="body_mass_g",y="flipper_length_mm",color='green')
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=penguins,x="body_mass_g",y="flipper_length_mm",hue="island",style ="sex")
sns.set_style("whitegrid")
```

```python
sns.jointplot(data=penguins,y='flipper_length_mm',x='body_mass_g')
```

```python
sns.jointplot(data=penguins,y='flipper_length_mm',x='body_mass_g',hue="sex",kind='kde')
```

```python
sns.barplot(data=penguins,x="species",y="body_mass_g",hue="sex",palette=['skyblue','pink'],estimator=np.sum)
```

```python
sns.countplot(data=penguins,x='species',hue='island')
```

```python
#box plot
sns.boxplot(data=penguins,x='species',y='body_mass_g',hue='sex')
```

```python
#vilonplot
sns.violinplot(data=penguins,x='species',y='body_mass_g',hue='sex')#,split=True)
```

```python
sns.violinplot(data=penguins,x='species',y='body_mass_g',palette='pastel')
sns.swarmplot(data=penguins,x='species',y='body_mass_g',color='black',hue="island",size=3,palette='dark')
```

```python
#KDE plot
sns.kdeplot(data=penguins,x='body_mass_g',hue='species',fill=True)
```

```python
#heat map

columns= ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]

penguins[columns].corr()
```

```python
sns.heatmap(penguins[columns].corr(),annot=True,cmap="Reds",vmin=-1,vmax=1)
```

```python
#rug plot
sns.rugplot(data=penguins,y="body_mass_g", hue="species",palette="deep",height=0.2)
```

```python
#pair plot
# sns.pairplot(data=penguins,hue="species",palette="pastel")
sns.pairplot(data=penguins)
```

```python
sns.pairplot(data=penguins,hue="sex",palette="pastel",diag_kind="kde")
```

```python
#pair f=grid
graph = sns.PairGrid(data=penguins, hue="sex", palette="pastel")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
